```python
Q1 = concat_data['weathersit'].quantile(0.25)
Q3 = concat_data['weathersit'].quantile(0.75)
IQR = Q3 - Q1
print("Q1: ", Q1, "\nQ3: ",Q3, "\nIQR:",IQR)


lower = Q1 - 1.5 * IQR
lower_array = concat_data['weathersit'] <= lower
print("Lower Bound:", lower)
print(lower_array.sum())

upper = Q3 + 1.5 * IQR
upper_array = concat_data['weathersit'] >= upper
print("Upper Bound:", upper)
print(upper_array.sum())

data_nooutliers1 = concat_data[~(lower_array | upper_array)]

print("Old Shape: ", concat_data.shape)
print("Old Shape: ", data_nooutliers1.shape)
```
